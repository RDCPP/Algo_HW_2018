```python
import matplotlib.pyplot as plt
x = [2,4,8,16,32,64,128,256,512,1024]
y1 = [4 * (n ** 2.81) for n in x]
y2 = [5 * (n ** 2.81) for n in x]
y3 = [6 * (n ** 2.81) for n in x]
y4 = [7 * (n ** 2.81) for n in x]
y = [7130,7667,31711,87437,464945,731628,5384990,37227536,254801833,1803108614]
plt.plot(x,y1,color='r',label = 'c = 4')
plt.plot(x,y2,color='g',label = 'c = 5')
plt.plot(x,y3,color='b',label = 'c = 6')
plt.plot(x,y4,color='y',label = 'c = 7')
plt.plot(x,y,color='black',label = 'c = strassen')
plt.legend()
plt.show()
```
